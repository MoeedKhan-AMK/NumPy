## NP.MAX IN NUMPY
NumPy's `max()` function is a powerful tool for array computation and analysis, used to find the maximum value within an array or along a specified axis. It is essential for data analysis tasks where determining the largest value in a dataset is necessary.

NumPy's max() function is used to identify the largest value in an array or along a specified axis. It helps pinpoint peak values that represent upper bounds, extremes, or highest-performing metrics in data.
np.max() returns the maximum element from the entire array or maximum values across rows/columns when axis is specified.

*MAX = Highest value in a dataset*
`Syntax: numpy.max(array, axis=None, output)`

## NP.MIN() IN NUMPY

The NumPy `min()` function is a part of Array Computation & Analysis tools, which is used to find the smallest value in an array. It is an efficient method for determining minimum values across various dimensions of a NumPy array.

NumPy's min() function returns the smallest value in an array or along a specific axis. It helps detect low-performance points, minimum readings, or worst-case values in datasets.
`np.min()`is used to compute the minimum numerical value across the entire dataset or per dimension.

*MIN = Lowest value in a dataset*


```python
# Finding Total Sales/Year OR sales per year.

yearly_total = np.sum(sales_data[:, 1:],axis=1)
print("Total 4 Years Sales Data of each Restaurant: :",yearly_total)

#Row-wise Sales Total:
restaurant_sales_total_each = np.sum(sales_data[0,1:], axis=0) #will exclude restaurant id, sum total years of sale.
print(f"Student Biryani total sales:",restaurant_sales_total_each)

#Use axis= 0 for column wise operation(goes top to bottom). Use axis=1 for row wise operation(goes left to right).

# Min sales for each restaurant:
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Minimum Sales/Restaurant:", min_sales)

# Max sales by each restaurants per year:
max_sales = np.max(sales_data[:, 1:], axis=0)
print("Maximum Sales Per Year:",max_sales)

# SIMPLE SUMMARY
# Operation 	 Meaning
# - axis=0	        Go down columns (per year)
# - axis=1	        Go across rows (per restaurant)
# - data[0, 1:]	    First row, ignore ID
# - data[:, 1:]	    All rows, ignore IDs
```
