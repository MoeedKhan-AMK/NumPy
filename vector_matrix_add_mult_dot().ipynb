## VECTOR MATRIX ADDITION - MULTIPLICATION - NP.DOT()

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Vector Matrix addition: 
We can add (or subtract) two matrices or two vectors, as long as the dimensions of both are the same.
![alt text](1_Ch1sWTQQB4WvkWyUVgz0tg.png)

### Vector Matrix Multiplication:
To perform vector-matrix multiplication, the number of columns in the matrix must equal the number of elements in the vector.
![alt text](multiplication.png)

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector Addition", vector1 + vector2)
print("\nVector Multiplication", vector1 * vector2)
```

## DOT PRODUCT NP.DOT() IN NUMPY:
- Linear Algebra is a branch of mathematics that deals with vectors, matrices, and operations on these entities.
- In NumPy, `the dot() function` is used to perform dot products of two arrays, which is fundamental in various mathematical computations including those in machine learning and data science.

`Syntax: NP.DOT(a,b)`

**Inputs	                        Meaning**
- `Two 1D arrays`	    Scalar dot product (sum of element-wise products)
- `Two 2D arrays`	    Matrix multiplication
- `1D & 2D array`	    Vector–matrix multiplication

**Real-world Use Cases**
- Machine Learning:	    *Weights × inputs in neural networks*
- NLP:	                *Cosine similarity of text embeddings*
- Physics:	            *Work done: force · displacement*
- Recommendation Systems:	    *User–item vector similarity*
- Computer Graphics	Vector:      *projections, transformations*

```python
# Dot Product of Vectors (COMMON)
print("Dot Product of Two Vector: ", np.dot(vector1, vector2))

#Dot Product of 2D ARRAY. Mtrix Multiplication
vector_1d = np.array([[1,2],[3,4]])
vector_2d = np.array([[10,20],[100, 200]])
print("\n Dot Product of Matrix Multiplication: \n", np.dot(vector_1d, vector_2d))

#VECTOR-MATRIX MULTIPLICATION:
vector = np.array([18,28])
matrix = np.array([[10,20],
                   [100,200]])

print("\n Vector Matrix Multiplication: ", np.dot(vector, matrix))


#Calculating Angle(Will Learn in detail later on.This Includee Linear Algebra. NOT IMP FOR NOW)
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\n Angle of Vectors (radians):", angle)
print("\n Angle of Vectors (degrees):", np.degrees(angle))
```
